```python
# 고객 성별-나이별 분류
tbl_demo.head()
```

```python
tbl_demo['ma_fem_dv_'] = [0 if i=='여성' else 1 for i in tbl_demo['ma_fem_dv']]
```

```python
tbl_demo['ages_'] = [int(i[:-1]) for i in tbl_demo['ages']]
tbl_demo.head()
```

```python
tbl_demo['ages_'].value_counts()
```

```python
df_by_group = tbl_demo.groupby(['ma_fem_dv_','ages_']).count()
df_by_group
```

```python
df_by_group = df_by_group.reset_index()
df_pivot = df_by_group.pivot(index='ages_',columns='ma_fem_dv_',values='cust')
df_pivot
```

```python
# https://12jeong.github.io/python-barplot/
df_pivot.plot.bar()
# 이쁘게 그리려면 https://jimmy-ai.tistory.com/40
```

성별 및 나이별 고객 수 결과
- 여성의 비율이 더 높음  
- 분포는 남녀 비슷하게 나타남

### 소현 - 고객별 어느 주기로 어떤 상품을 구매했는지 조사 - 0629까지

```python
# 구매 내역 상위 100을 토대로 우선 조사 -> 고객번호와 영수증 구매금액을 토대로 상위 100을 선정
tbl_pdde.head()
# 여기서 질문 하나: 구매금액 - 구매수량이면 총 구매금액은 구매금액*구매수량인가?
```

```python
df_top = tbl_pdde.groupby(['cust'])['buy_am'].sum()
df_top = df_top.reset_index()
df_top
# 여기서 알게 된 사실 중 하나: 고객수는 29913인데 실제 구매한 내역이 있는 고객수는 26917
```

```python
df_top=df_top.sort_values('buy_am',ascending=False)[:100]
df_top
```

```python
# 이 100명에 대한 구매 정보를 다시 불러올 것임
tbl_top=tbl_pdde[tbl_pdde['cust'].isin(df_top['cust'].tolist())]
tbl_top
```

```python
# 구매금액 상위 100 고객에 대한 특징 분석: 성별-나이별 분석
tbl_top_des = tbl_demo[tbl_demo['cust'].isin(df_top['cust'].tolist())]
tbl_top_des
```

```python
df_top_des = tbl_top_des.groupby(['ma_fem_dv_','ages_']).count()
df_top_des = df_top_des.reset_index()
df_pivot_top = df_top_des.pivot(index='ages_',columns='ma_fem_dv_',values='cust')
df_pivot_top
```

```python
df_pivot_top = df_pivot_top.fillna(0)
df_pivot_top
```

```python
df_pivot_top.plot.bar()
# 상위 50% 고객 수 / 일반 고객 수의 비율을 보는 것도 재밋을듯
```

```python
# 고객별 어느 주기로 어떤 상품을 구매했는지 조사(구매금액 상위 100에 대하여)
pd.merge(left=tbl_top,right=tbl_pd_clac,how='left',on='pd_c')
# pd_c가 상품 코드인데 tbl_pdde에서는 점포코드로 표기되어있음.
```

```python
df_top_clac = pd.merge(left=tbl_top,right=tbl_pd_clac,how='left',on='pd_c')
df_top_clac = df_top_clac[['cust','rct_no','pd_nm','clac_hlv_nm','clac_mcls_nm']]
df_top_clac
```

```python
# clac_mcls_nm별 count
df_top_clac = df_top_clac.groupby(['clac_mcls_nm'])['cust'].count()
df_top_clac = df_top_clac.reset_index()
df_top_clac
```

```python
df_top_clac.sort_values('cust',ascending=False)
```

```python
# 한글안깨지게 - https://seonghyuk.tistory.com/31
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = 'NanumGothic'

df_top_clac_20 = df_top_clac.sort_values('cust',ascending=False)[:20]
df_top_clac_20 = df_top_clac_20.set_index('clac_mcls_nm')
df_top_clac_20.plot.bar()
```
